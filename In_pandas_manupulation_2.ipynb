```python
df['new_col_square_d']=df['d'].apply(lambda x:x**2)
```

```python
df['new_col_log_d']=df['d'].apply(lambda x:np.log(x))
```

```python
df
```

```python
# -ve value keep as same and the log for the positive number 
def test(x):
    if (x<0):
        return x
    else:
        return np.log(x)
        
```

```python
df['new_col_log_d1']=df['d'].apply(test)
```

```python
df
```

```python
def test1(x):
    if x <0:
        return 1
    else:
        return 0
```

```python
df['new_a']=df['a'].apply(test1)
```
