#### Multiple lines on the same plot

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,11)
i= x
s= x**2
c= x**3
plt.plot(x,i,'o-r')
plt.plot(x,s,'o-b')
plt.plot(x,c,'o-g')
plt.title("One graph, but multiplot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,11)
i= x
s= x**2
c= x**3
plt.plot(x,i,'o-r',x,s,'o-b',x,c,'o-g')
plt.title("One graph, but multiplot")
plt.show()
```

#### How to set the limits range of values on x-axis and y-axis
1. xlim()-(left,right)
2. ylim()-(bottom,top)

```python
help(plt.xlim)
```

```python
# when we are not giving any limit it will take the default value provided by pvm
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(1,101)
b= a**2
plt.plot(a,b,'o-r')
plt.grid()
left,right=plt.xlim()
bottom,top=plt.ylim()
print("left limit on the x-axis",left)
print("right limit on the x-axis",right)
print("bottom limit on the y-axis",bottom)
print("top limit on the x-axis",top)
plt.show()
```

#### 3. plotting the graphs and setting axes properties

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(1,11)
b = a**2
fig=plt.figure()# figure object created at the mentioned location
axes=fig.add_axes([0.2,0.3,0.6,0.4])#The dimensions [left, bottom, width, height] of the new axes
axes.plot(a,b)
axes.set_xlabel("xlabel")
axes.set_ylabel("ylabel")
axes.set_title("title")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(1,11)
b = a**2
fig=plt.figure()# figure object created at the mentioned location
axes=fig.add_axes([0.2,0.3,0.6,0.4])#The dimensions [left, bottom, width, height] of the new axes
axes.plot(a,b)
axes.set_xlabel("N")
axes.set_ylabel("Square of N")
axes.set_title("Square function")
plt.show()
```

```python
# in short form
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(1,11)
b = a**2
fig=plt.figure()# figure object created at the mentioned location
axes=fig.add_axes([0.2,0.3,0.6,0.4])#The dimensions [left, bottom, width, height] of the new axes
axes.plot(a,b)
axes.set(xlabel="N",ylabel="Square of N",title="Square of Function",xlim=(1,5),ylim=(1,25))

plt.show()
```
