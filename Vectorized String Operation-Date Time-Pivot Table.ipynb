#### Pivot Table
  The pivot table takes simple column-wise data as input,and group the entries into a two-dimensional table that provides a multidimensional summarization of tha data.

```python
df=sns.load_dataset("tips") # a default dataset(tips -> a resturant dataset ) comes with seaborn library 
```

```python
df.head()
```

```python
df.shape
```

```python
# get the average of the male/female total_bill
```

```python
df.groupby("sex")[["total_bill"]].mean()
```

```python
# the average bill paymnet with somker or non-smoker male or female
df.groupby(["sex","smoker"])[["total_bill"]].mean()# multiindex dataframe 
```

```python
df.groupby(["sex","smoker"])[["total_bill"]].mean().unstack()# multiindex dataframe 
```

```python
#  pivot table
df.pivot_table(index="sex",columns="smoker",values="total_bill")
```

```python
# here the aggrigate function by default is mean()
```

```python
# aggfunc
#  sum/std/var/userdefine mathematical function
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="var")
```

```python
# all columns together
#  pivot table
df.pivot_table(index="sex",columns="smoker")
# if we not provide the value it will do analysis of all the numerical columns in the dataframe 
```

```python
df.pivot_table(index="sex",columns="smoker")["size"]
```

```python
# multidimensional 
df.pivot_table(index=["sex","smoker"],columns=["day","time"],values="total_bill")
```

```python
# we can also find the different different agg function in diffferent columns 
```

```python
# size->mean,tip->max,total_bill->sum
df.pivot_table(index=["sex","smoker"],columns=["day","time"],aggfunc={"size":"mean","tip":"max","total_bill":"sum"})
```

```python
#margins-> the total of tha dataframe 
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum")
```

```python
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum",margins=True)
```
