### Series is 1D and DataFrames are 2D objects

- But why?
   To fetch the data from series we need only one item, but to fetch data from dataframe we need two item.
- And what exactly is index?

```python
# can we have multiple index?
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
```

```python
index_val
```

```python
a=pd.Series([1,2,3,4,5,6,7,8],index=index_val)
```

```python
a
```

```python
a[('cse',2020)]
```

```python
a[("cse")] # we can't find the cse alone
```

```python
# The solution -> multiindex series (also known as hierrachical Indexing)
# multiple indexlevel within a single index
```

```python
# how to create multiindex object
# 1. pd.MultiIndex.from_tuples() 
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex=pd.MultiIndex.from_tuples(index_val)
multiindex
```

```python
multiindex.levels
```

```python
multiindex.levels[0]
```

```python
multiindex.levels[1]
```

```python
# 2. pd.MultiIndex.from_product()
pd.MultiIndex.from_product([["cse","ece"],[2019,2020,2021,2022]])
```

```python
# creating Series with multiindex object
s=pd.Series([1,2,3,4,5,6,7,8],index=multiindex)
s
```

```python
# fetching item from multiindexed series
s[("cse",2019)] # for single item 
```

```python
s["cse"] # item for cse branch
```

```python
#unstack -> we can convert a multiindex series into dataframe
```

```python
s
```

```python
s.unstack()
```

```python
temp=s.unstack()
temp
```

```python
#stack()-> dataframe to multiindex series
temp.stack()
```
