```python
import matplotlib.pyplot as plt
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
sales=[10000,25000,45000,30000,10000,5000,70000,60000,65000,50000]
c=['r','k','b','g','orange','yellow','m','c','lime','violet']
plt.bar(year,sales,color=c)
plt.xlabel('Year',color='r',fontsize=15)
plt.ylabel('Number of Sales',color='r',fontsize=15)
plt.title('Nokia mobile sales in last decade',color='r',fontsize=15)
plt.xticks(year,rotation=30)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

#### How to add label to the bars 


1. text()
2. annotate()

```python
help(plt.text)
```

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.arange(10)
plt.plot(a,a,'o-r')
for i in range(a.size):
    plt.text(a[i]+0.4,a[i]-0.1,f'({a[i],a[i]})',color='b')
    
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.arange(10)
plt.plot(a,a,'o-r')
for i in range(a.size):
    plt.annotate(f'({a[i],a[i]})',(a[i]+0.4,a[i]+0.2),color='b')
    
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
names= np.array([],dtype='str')
marks= np.array([],dtype='int')

f= open('C:\\Users\\91993\\Desktop\\student.csv','r')
r= csv.reader(f)
h= next(r)
for row in r :
    names = np.append(names,row[0])
    marks = np.append(marks,row[1])
plt.barh(names,marks,color='r')
plt.xlabel("marks",fontsize=15,color='b')
plt.ylabel("name",fontsize=15,color='b')
plt.title("Student performance",fontsize=15,color='k')
plt.tight_layout()
plt.show()
```

### 3. Stacked bars 

```python
names = ['Sunny','Bunny','Chinny','Vinny','Tinny']
english_marks = [90,80,85,25,50]
maths_marks = [25,23,45,32,50]
plt.bar(names,english_marks,color="green")
plt.bar(names,maths_marks,bottom=english_marks,color='r') # putting the bottom as the englsh_marks
```

```python
import matplotlib.pyplot as plt
import numpy as np
country_name = ['India','China','US','UK']
gold_medals = np.array([60,40,50,20])
silver_medals = np.array([50,30,25,43])
bronze_medals = np.array([55,24,45,6])
plt.bar(country_name,gold_medals,color="y")
plt.bar(country_name,silver_medals,bottom=gold_medals,color='gray')
plt.bar(country_name,bronze_medals,bottom=gold_medals+silver_medals,color='maroon')
plt.xlabel('Country Name',color='b',fontsize=15)
plt.ylabel('Number of Medals',color='b',fontsize=15)
plt.title('Country Wise Medals Report',color='r',fontsize=15)
plt.legend(['gold','silver','bronze'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
country_name = ['India','China','US','UK']
gold_medals = np.array([60,40,50,20])
silver_medals = np.array([50,30,25,43])
bronze_medals = np.array([55,24,45,6])
plt.bar(country_name,gold_medals,color="#ffd700")
plt.bar(country_name,silver_medals,bottom=gold_medals,color='#c0c0c0')
plt.bar(country_name,bronze_medals,bottom=gold_medals+silver_medals,color='#cd7f32')
plt.xlabel('Country Name',color='b',fontsize=15)
plt.ylabel('Number of Medals',color='b',fontsize=15)
plt.title('Country Wise Medals Report',color='r',fontsize=15)
plt.legend(['gold','silver','bronze'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
country_name = ['India','China','US','UK']
gold_medals = np.array([60,40,50,20])
silver_medals = np.array([50,30,25,43])
bronze_medals = np.array([55,24,45,6])
plt.barh(country_name,gold_medals,color="#ffd700")
plt.barh(country_name,silver_medals,left=gold_medals,color='#c0c0c0')
plt.barh(country_name,bronze_medals,left=gold_medals+silver_medals,color='#cd7f32')
plt.ylabel('Country Name',color='b',fontsize=15)
plt.xlabel('Number of Medals',color='b',fontsize=15)
plt.title('Country Wise Medals Report',color='r',fontsize=15)
plt.legend(['gold','silver','bronze'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
names = ['Sunny','Bunny','Chinny','Vinny','Tinny']
english_marks = np.array([90,80,85,25,50])
math_marks = np.array([25,23,45,32,25])
total_marks = english_marks+math_marks
plt.bar(names,english_marks,color='#09695c',label='English')
plt.bar(names,math_marks,bottom=english_marks,color='#9c0c8b',label="Maths")

for i in range(len(names)):
    plt.text(names[i],(english_marks[i]/2),str(english_marks[i]),ha='center',color='white',weight=1000)
    plt.text(names[i],(english_marks[i]+math_marks[i]/2),str(math_marks[i]),ha='center',color='white',weight=1000)
    plt.text(names[i],(total_marks[i]+2), str(total_marks[i]),ha='center',color='#008080',weight=1000)
    
plt.xlabel("Name of the Student",color='#570b66', fontsize=15,weight=1000)
plt.ylabel("Marks",color='#570b66', fontsize=15,weight=1000)
plt.title("Student -wise Marks",color='#ED0A3F', fontsize=15,weight=1000)
plt.legend()
plt.tight_layout()
plt.show()
```
